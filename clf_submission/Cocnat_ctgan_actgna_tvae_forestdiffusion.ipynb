```python
import pandas as pd

# 모델별 CSV 파일 경로 
file_paths = {
    'CTGAN': "/data/CTGAN_clf_submission.csv",
    'ACTGAN': "/data/ACTGAN_clf_submission.csv",
    'ForestDiffusion': "/data/ForestDiffusion_clf_submission.csv",
    'TVAE': "/data/TVAE_clf_submission.csv",
    'Train': "../data/train.csv"
}

# 각 Fraud_Type에 대해 사용할 모델 매핑 (m은 train.csv에서 가져옴)
fraud_type_model_mapping = {
    'a': 'CTGAN',
    'b': 'ForestDiffusion',
    'c': 'ForestDiffusion',
    'd': 'CTGAN',
    'e': 'TVAE',
    'f': 'ForestDiffusion',
    'g': 'CTGAN',
    'h': 'ACTGAN',
    'i': 'ForestDiffusion',
    'j': 'CTGAN',
    'k': 'CTGAN',
    'l': 'CTGAN',
    'm': 'Train'  
}


final_data = pd.DataFrame()



# 각 모델에 대해 데이터 로드 및 결합
for fraud_type, model_name in fraud_type_model_mapping.items():
    
    model_data = pd.read_csv(file_paths[model_name])
    print(model_data)
    print(model_data.shape)
    
    # 해당 Fraud_Type의 데이터만 필터링
    fraud_type_data = model_data[model_data['Fraud_Type'] == fraud_type]
    
    # 최종 데이터프레임에 결합
    final_data = pd.concat([final_data, fraud_type_data], ignore_index=True)

# 최종 데이터를 CSV 파일로 저장
final_data.to_csv('../data/others.csv', encoding='UTF-8-sig', index=False)

print("Data processing complete. Combined data saved to 'combined_fraud_type_data.csv'.")
```
